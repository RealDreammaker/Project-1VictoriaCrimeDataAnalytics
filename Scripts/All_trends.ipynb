```python
# Criminal Incidents by year - police service regions comparision
lga_ci_tab1_nth = lga_ci_tab1['Incidents Recorded'].loc[(lga_ci_tab1['Police Region'].str[:1] == '1') 
                                                       & (lga_ci_tab1['Local Government Area'] == 'Total')].values
lga_ci_tab1_sth = lga_ci_tab1['Incidents Recorded'].loc[(lga_ci_tab1['Police Region'].str[:1] == '3') 
                                                       & (lga_ci_tab1['Local Government Area'] == 'Total')].values
lga_ci_tab1_east = lga_ci_tab1['Incidents Recorded'].loc[(lga_ci_tab1['Police Region'].str[:1] == '2') 
                                                       & (lga_ci_tab1['Local Government Area'] == 'Total')].values
lga_ci_tab1_west = lga_ci_tab1['Incidents Recorded'].loc[(lga_ci_tab1['Police Region'].str[:1] == '4') 
                                                       & (lga_ci_tab1['Local Government Area'] == 'Total')].values

# Get Crime incidents rate per 100K population by Police Service Areas Regions
lga_ci_tab1_rate1 = lga_ci_tab1['Rate per 100,000 population'].loc[(lga_ci_tab1['Police Region'].str[:1] == '1') 
                                                       & (lga_ci_tab1['Local Government Area'] == 'Total')].values
lga_ci_tab1_rate2 = lga_ci_tab1['Rate per 100,000 population'].loc[(lga_ci_tab1['Police Region'].str[:1] == '2') 
                                                       & (lga_ci_tab1['Local Government Area'] == 'Total')].values
lga_ci_tab1_rate3 = lga_ci_tab1['Rate per 100,000 population'].loc[(lga_ci_tab1['Police Region'].str[:1] == '3') 
                                                       & (lga_ci_tab1['Local Government Area'] == 'Total')].values
lga_ci_tab1_rate4 = lga_ci_tab1['Rate per 100,000 population'].loc[(lga_ci_tab1['Police Region'].str[:1] == '4') 
                                                       & (lga_ci_tab1['Local Government Area'] == 'Total')].values

# Data cleaning - drop duplicates to unique year's value
ci_year = lga_ci_tab1['Year'].drop_duplicates()

# Create new dataframe to hold the required data for each police service area region
ci_regional_sum =[]
ci_regional_sum = pd.DataFrame({"Year": ci_year})
ci_regional_sum["North West Criminal Incidents"] = lga_ci_tab1_nth
ci_regional_sum["South Criminal Incidents"] = lga_ci_tab1_sth
ci_regional_sum["East Criminal Incidents"] = lga_ci_tab1_east
ci_regional_sum["West Criminal Incidents"] = lga_ci_tab1_west
ci_regional_sum["Rate All Regions per 100k pop"] = lga_ci_tab1_rate1 + lga_ci_tab1_rate2 + lga_ci_tab1_rate3 + lga_ci_tab1_rate4
```

```python
ci_regional_sum = ci_regional_sum.sort_values(by="Year").reset_index(drop=True)
ci_regional_sum 
```

```python
ci_regional_sum = ci_regional_sum.sort_values(by="Year")

# Set up graph information
fig, ax1 = plt.subplots(figsize = (15,8))
ax2 = ax1.twinx()
ax2.set_ylabel('Rate per 100K Population',fontsize=12)
ax1.set_ylabel('Number of Incidents',fontsize=12)

# Generate bars plot showing the distribution crime incidents for each region
ci_regional_sum.plot(x="Year",y=["North West Criminal Incidents","South Criminal Incidents","East Criminal Incidents","West Criminal Incidents"],kind="bar",ax=ax1)
ci_regional_sum.plot(y=["Rate All Regions per 100k pop"],kind="line", color='red',ax=ax2)

# Relocate the legend box outside the graph
ax1.legend(bbox_to_anchor=(1.04,1), loc="upper left")
ax2.legend(bbox_to_anchor=(1.04,0.8), loc="upper left")

# Label & plot graphs
plt.title("Criminal Incidents",fontsize=16)
plt.xlabel("Year",fontsize=14)
plt.show()
```
